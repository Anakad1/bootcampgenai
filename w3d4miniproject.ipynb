```python
# Install required libraries
!pip install wget yfinance scipy

# Import required modules
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import seaborn as sns
import os
import wget
import yfinance as yf
from google.colab import files
from IPython.display import display

# Upload file from the user's local machine
uploaded = files.upload()

# Check if any files were uploaded
if len(uploaded) > 0:
    # Get the first file name uploaded
    first_file_name = list(uploaded.keys())[0]

    # Check if the file name ends with .csv, indicating it is a csv file
    if first_file_name.endswith('.csv'):
        # Read the uploaded file into a pandas DataFrame
        df = pd.read_csv(first_file_name)

        # Check if 'Date' column exists
        if 'Date' in df.columns:
          #Display the first few rows of the dataframe to verify
          display(df.head())
        else:
          print("The uploaded file does not contain a 'Date' column.")
          print("Attempting to retrieve AAPL data from Yahoo Finance instead")
          # Retrieve AAPL data from Yahoo Finance
          df = yf.download('AAPL', start='2013-01-01', end='2023-12-31')
          # Reset the index to make Date a column
          df.reset_index(inplace=True)
          display(df.head())
          print('Successfully loaded data from yfinance')
    else:
      print("The uploaded file is not a csv file.")
      print("Attempting to retrieve AAPL data from Yahoo Finance instead")
      # Retrieve AAPL data from Yahoo Finance
      df = yf.download('AAPL', start='2013-01-01', end='2023-12-31')
      # Reset the index to make Date a column
      df.reset_index(inplace=True)
      display(df.head())
      print('Successfully loaded data from yfinance')
else:
  print("No files were uploaded.")
  print("Attempting to retrieve AAPL data from Yahoo Finance instead")
  # Retrieve AAPL data from Yahoo Finance
  df = yf.download('AAPL', start='2013-01-01', end='2023-12-31')
  # Reset the index to make Date a column
  df.reset_index(inplace=True)
  display(df.head())
  print('Successfully loaded data from yfinance')

# Display null values in the dataframe.
display(df.isnull().sum())

# Plot closing price over time
plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Close'], label='Close Price', color='blue')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('AAPL Closing Prices Over Time')
plt.legend()
plt.grid()
plt.show()

# Plot trading volume over time
plt.figure(figsize=(12,6))
plt.bar(df['Date'], df['Volume'], label='Volume', color='orange')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('AAPL Trading Volume Over Time')
plt.legend()
plt.grid()
plt.show()
```

```python
stats_summary = df[['Open', 'High', 'Low', 'Close', 'Volume']].describe()
display(stats_summary)
```

```python
# Calculate moving average
df['MA_50'] = df['Close'].rolling(window=50).mean()
plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Close'], label='Close Price', color='blue')
plt.plot(df['Date'], df['MA_50'], label='50-Day Moving Avg', color='red')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('AAPL Closing Prices with 50-Day Moving Average')
plt.legend()
plt.grid()
plt.show()
```

```python
# Compute daily returns
df['Daily_Return'] = df['Close'].pct_change()
plt.figure(figsize=(12,6))
sns.histplot(df['Daily_Return'].dropna(), bins=50, kde=True, color='purple')
plt.title('Distribution of Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
year_2022 = df[df['Date'].dt.year == 2022]['Close']
year_2023 = df[df['Date'].dt.year == 2023]['Close']
t_stat, p_value = stats.ttest_ind(year_2022.dropna(), year_2023.dropna(), equal_var=False)
print(f'T-Test Statistic: {t_stat}, P-Value: {p_value}')
```

```python
df['MA_20'] = df['Close'].rolling(window=20).mean()
df['MA_100'] = df['Close'].rolling(window=100).mean()
correlation = np.corrcoef(df['MA_20'].dropna(), df['Volume'].dropna())[0,1]
print(f'Correlation between 20-day moving average and volume: {correlation}')
```
