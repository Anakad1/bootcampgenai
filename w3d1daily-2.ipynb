```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
num_cities = 10
num_months = 12
```

```python
temperature_data = np.random.uniform(low=-5, high=35, size=(num_cities, num_months))
```

```python
cities = [f"Ville_{i+1}" for i in range(num_cities)]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
```

```python
df_temperatures = pd.DataFrame(temperature_data, index=cities, columns=months)
print("DataFrame des températures mensuelles :")
print(df_temperatures)
```

```python
annual_avg = df_temperatures.mean(axis=1)
print("\nTempérature moyenne annuelle pour chaque ville :")
print(annual_avg)
```

```python
ville_max = annual_avg.idxmax()
ville_min = annual_avg.idxmin()
print(f"\nLa ville avec la température moyenne la plus élevée : {ville_max}")
print(f"La ville avec la température moyenne la plus basse : {ville_min}")
```

```python
plt.figure(figsize=(10, 6))
annual_avg.plot(kind='bar', color='skyblue')
plt.title("Température moyenne annuelle par ville")
plt.xlabel("Ville")
plt.ylabel("Température moyenne (°C)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for city in df_temperatures.index:
    plt.plot(months, df_temperatures.loc[city], marker='o', label=city)
plt.title("Températures mensuelles pour chaque ville")
plt.xlabel("Mois")
plt.ylabel("Température (°C)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
