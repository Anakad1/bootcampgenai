```python
import pandas as pd

# Charger les données
df = pd.read_csv("datascience_salaries.csv")

# Afficher les premières lignes pour vérifier le contenu
df.head()
```

```python
# Vérifier les premières lignes
print(df.head())

# Vérifier les infos du dataset
print(df.info())

# Vérifier les statistiques générales
print(df.describe())
```

```python
# Regrouper par niveau d'expérience et calculer le salaire moyen et médian
salary_stats = df.groupby("experience_level")["salary"].agg(["mean", "median"]).reset_index()

# Renommer les colonnes pour plus de clarté
salary_stats.columns = ["experience_level", "mean_salary", "median_salary"]

# Afficher les statistiques de salaire par niveau d'expérience
print(salary_stats)
```
