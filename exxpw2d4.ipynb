La visualisation des données est cruciale car elle permet de transformer des données brutes et souvent complexes en images claires et compréhensibles.
Un graphique linéaire est principalement utilisé pour montrer l’évolution d’une variable sur une échelle continue, souvent au fil du temps. Il met en évidence des tendances, des augmentations ou des diminutions progressives

exercice 2


```python
import matplotlib.pyplot as plt

#températures sur une semaine
temperatures = [72, 74, 76, 80, 82, 78, 75]
days = ["Lundi", "Mardi", "Mercredi", "Jeudi", "Vendredi", "Samedi", "Dimanche"]

#création du graphique linéaire
plt.plot(days, temperatures, marker='o', linestyle='-', color='blue')
plt.xlabel("Jour")
plt.ylabel("Température (°F)")
plt.title("Variation de la température sur une semaine")
plt.show()
```

```python
#ventes mensuelles
sales = [5000, 5500, 6200, 7000, 7500]
months = ["Janvier", "Février", "Mars", "Avril", "Mai"]

#création du graphique à barres
plt.bar(months, sales, color='orange')
plt.xlabel("Mois")
plt.ylabel("Montant des ventes ($)")
plt.title("Ventes mensuelles")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (replace with your actual data)
data = {
    "Choisissez votre sexe": ["Homme", "Femme", "Homme", "Femme", "Homme", "Femme", "Homme", "Femme", "Homme", "Femme"],
    "Souffrez-vous d'anxiété ?": ["Oui", "Non", "Non", "Oui", "Oui", "Non", "Oui", "Non", "Non", "Oui"],
}
df = pd.DataFrame(data)  # Create DataFrame from sample data

# Graphique à barres comparant les niveaux d'anxiété entre les sexes
sns.countplot(
    data=df,
    x="Choisissez votre sexe",
    hue="Souffrez-vous d'anxiété ?",
    palette="pastel",
)
plt.title("Comparaison des niveaux d'anxiété entre les sexes")
plt.xlabel("Sexe")
plt.ylabel("Nombre d'étudiants")
plt.legend(title="Souffrez-vous d'anxiété ?", loc="upper right")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (replace with your actual data)
data = {
    "Âge": [25, 30, 22, 35, 28, 32, 27, 31, 29, 33],  # Added 'Âge' column
    "Avez-vous des crises de panique ?": ["Oui", "Non", "Non", "Oui", "Oui", "Non", "Oui", "Non", "Non", "Oui"],
}
df = pd.DataFrame(data)  # Create DataFrame from sample data

# Convertir les réponses aux attaques de panique en valeurs numériques
df["Crises de panique (numériques)"] = df["Avez-vous des crises de panique ?"].map({"Oui": 1, "Non": 0})

# Nuage de points montrant la relation âge vs crises de panique
sns.scatterplot(data=df, x="Âge", y="Crises de panique (numériques)", alpha=0.6, s=100)
plt.title("Relation entre l'âge et les crises de panique")
plt.xlabel("Âge")
plt.ylabel("Présence de crises de panique")
plt.show()
```
